```python
import pandas as pd
```

```python
#data dict should help get columns for each assessment

data_dict = pd.read_csv('Voice_DataDictionary_2022-02-25.csv', index_col=0, encoding="iso-8859-1")
data_dict
```

```python
form_list = data_dict['Form Name'].unique()
form_list
```

```python
pheno = pd.read_csv('../../participants.tsv', sep='\t', index_col=0)
```

```python
def find_column_names(data_dict, form_name):
    
    #input data dict. Returns columns with the name of the given form
    
    #this returns the construct tested by the question
    #return data_dict['Field Label'][data_dict['Form Name'] == form_name] 
    
    return list(data_dict[data_dict['Form Name'] == form_name].index)
```

```python
split = {}

for form_name in form_list:
    col = find_column_names(data_dict, form_name)
    
    if form_name == 'initial_screening':
        #these columns were removed to deidentify data so do not exist
        id_col = ['study_id', 'voice_id', 'mobile_id', 'first_name', 'last_name', 
                  'address', 'telephone_1', 'telephone_2', 'email', 'dob', 'live_in_boston_area']

        [col.remove(x) for x in id_col]
    elif form_name == 'qidssr16':
        #This column not in the phenotype spreadsheet
        col.remove('qidssr_instructions')
    
    #remove columns that calculated totals since they wrong
    col = [c for c in col if 'total' not in c] 

    #remove date and time columns because not needed
    col = [c for c in col if 'date' not in c] 
    col = [c for c in col if 'time' not in c] 
    
    split[form_name] = pheno[['participant_id'] + col]
```
