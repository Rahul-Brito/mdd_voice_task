```python
### creating long-form sum-score version

pheno_tables = [pd.read_table(a,  index_col=0) for a in assesments]
sum_score_tables = []
for pt in pheno_tables:
    pt.index = pt.participant_id
    pt.drop(columns='participant_id', axis=1, inplace=True)
    s = pt.sum(axis=1, min_count=1)# should make NaN sum to NaN not 0
    #s = s.dropna()
    s = s/s.max() #make between 0 and 1 to compare tests
    s.rename(pt.columns[0].split("_1")[0], inplace = True)
    sum_score_tables.append(s)

age_sex =  pd.read_table('../../phenotype/initial_screening.tsv')[['participant_id','sex', 'age']]
age_sex.index = age_sex.participant_id
age_sex.age = age_sex.age - age_sex.age.mean()
age_sex.drop(columns='participant_id', axis=1, inplace=True)


sum_score_tables.append(age_sex)
sum_scores_df = pd.concat(sum_score_tables, axis=1) #include sex and age as covaraites
sum_scores_df = sum_scores_df[~((sum_scores_df.index == 'sub-voice955') & (sum_scores_df.age == 53.0) & (sum_scores_df.sex == 'male'))] #remove redundant 955
sum_scores_df.drop('sub-voicenan', inplace=True)
#sum_scores_df.drop(columns=['age', 'sex'], inplace=True)

#remove people with NAN in the BDI including people with question scores >3
#bdi_df = pheno.loc[:,pheno.columns.str.contains('beckdepression')]
#pheno = pheno.loc[bdi_df[bdi_df < 3].dropna().index,:]
#pheno = pheno.loc[:,~pheno.columns.duplicated()]
#pheno = pheno.reset_index(drop=True)
sum_scores_df['participant_id'] = sum_scores_df.index
#sum_scores_df.reset_index()
sum_scores_df
```

```python
pheno_long_form = pd.melt(sum_scores_df, id_vars=['participant_id', 'age', 'sex'], value_vars = sum_scores_df.columns[:-1],
                          var_name= 'mh_scale', value_name= 'sum_score')
pheno_long_form = pheno_long_form.dropna()
pheno_long_form
```

```python
pheno_long_form.to_csv('PSY1950_FINAL_LONGFORM_mental_health_dataset.csv')
```
